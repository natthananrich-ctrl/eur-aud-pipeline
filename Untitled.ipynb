```python
import os

# 1) Confirm where you are
print("CWD:", os.getcwd())

# 2) Create the folders
target_dir = os.path.join(".github", "workflows")
os.makedirs(target_dir, exist_ok=True)
print("Created:", target_dir)

# 3) Write ci.yml content
ci_content = """name: CI Pipeline

on:
  push:
    branches: [ main ]
  pull_request:
    branches: [ main ]

jobs:
  build-and-test:
    runs-on: ubuntu-latest

    steps:
    - name: Checkout repo
      uses: actions/checkout@v3

    - name: Set up Python
      uses: actions/setup-python@v4
      with:
        python-version: '3.10'

    - name: Install dependencies
      run: |
        pip install -r requirements.txt

    - name: Run unit tests
      run: |
        pytest tests/ --disable-warnings --maxfail=1

    - name: Run pipeline
      run: |
        python cli/run_pipeline.py prod
"""

ci_path = os.path.join(target_dir, "ci.yml")
with open(ci_path, "w", encoding="utf-8") as f:
    f.write(ci_content)

print("Written:", ci_path)
```
